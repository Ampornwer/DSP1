```python
#ประก่าศเพื่อใช้งาน
from __future__ import print_function, division

%matplotlib inline

import thinkdsp
import thinkplot

import numpy as np

from ipywidgets import interact, interactive, fixed
import ipywidgets as widgets
from IPython.display import display

```

```python
#Labe1
sin_wave=thinkdsp.SinSignal(freq=7023, amp=0.5, offset=0)
sin_wave.plot()
thinkplot.config(xlabel='Time (s)')
```

```python
#Labe2
signal = thinkdsp.TriangleSignal(4000)
duration = signal.period*2500
segment = signal.make_wave(duration, framerate=10000)
segment.plot()
wave=signal.make_wave(duration, framerate=10000)
wave.apodize()
wave.make_audio()
spectrum = wave.make_spectrum()
spectrum.plot()
```

```python
signal = thinkdsp.SawtoothSignal(23)
duration = signal.period*23
segment = signal.make_wave(duration, framerate=10000)
segment.plot()
thinkplot.config(ylim=[-1.05, 1.05], legend=False)
segment.plot()
wave = signal.make_wave(duration=0.5, framerate=10000)
wave.apodize()
wave.make_audio()
```
